```python
dic=['PersonId','Name', 'LastName', 'CurrentRole','Country','Industry','NumberOfRecommendations','NumberOfConnections']
#using numpy's randint
df = pd.DataFrame(np.random.randint(4000,5000,size=(100, 8)), columns=dic)
df['Name'] = 'Gabriela'
df ['LastName']='Castillo Areco'
df ['CurrentRole']='Desarrollador'
df['Country']='Argentina'
df ['Industry'] = 'Turismo'
df ['Connect'] = 'Connect'
df.head(5)
```

```python
#https://faker.readthedocs.io/en/master/locales/es_ES.html#faker-providers-person
#https://faker.readthedocs.io/en/master/providers/faker.providers.lorem.html
from faker import Faker
#Name column
fake = Faker('es_ES')
for index, row in df.iterrows():
      df.at[index,'Name'] = fake.first_name()
#Last name column
for index, row in df.iterrows():
      df.at[index,'LastName'] = fake.last_name()
        
#CurrentRole column      
fake = Faker()
my_word_list = [
'developer','desarrollador','cto',
'ciencias de la computacion','web developer','computacion',
'data scientist','cientifico de datos','software developer',
'software engineer','desarrollador de software']     
for index, row in df.iterrows():
      df.at[index,'CurrentRole'] = fake.sentence(nb_words=1, ext_word_list=my_word_list)      
        
#Country column
my_word_list = [
'Antigua y Barbuda', 'Argentina', 'Bahamas', 'Barbados', 'Belice', 'Bolivia', 'Brasil', 'Colombia', 'Costa Rica', 'Cuba',
'Chile', 'Dominica', 'Ecuador', 'El Salvador', 'Granada', 'Guatemala', 'Guyana', 'Haiti', 'Honduras', 'Jamaica', 'M칠xico', 
'Nicaragua', 'Panama', 'Paraguay', 'Peru', 'Republica Dominicana', 'San Cristobal y Nieves', 'San Vicente y las Granadinas', 'Santa Luc칤a',
'Surinam', 'Trinidad y Tobago', 'Uruguay', 'Venezuela']     
for index, row in df.iterrows():
      df.at[index,'Country'] = fake.sentence(nb_words=1, ext_word_list=my_word_list)      
        
#Industry        
my_word_list = [
'Telecommunications','Network Security','telecomunicaciones',
'ciberseguridad','tecnologia','computacion',
'marketing','analitica','ventas']    
for index, row in df.iterrows():
      df.at[index,'Industry'] = fake.sentence(nb_words=1, ext_word_list=my_word_list)   

#NumberOfRecommendations - idealmente usar random_digit_not_null_or_empty()
#https://faker.readthedocs.io/en/master/providers/baseprovider.html
for index, row in df.iterrows():
      df.at[index,'NumberOfRecommendations'] = random.randint(0,10)

 ##NumberOfConnections
for index, row in df.iterrows():
      df.at[index,'NumberOfConnections'] = random.randint(0,500)
df
```
